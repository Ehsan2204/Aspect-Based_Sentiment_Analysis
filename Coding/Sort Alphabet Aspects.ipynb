```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('dataset_modelling2.csv')

# Function to sort aspect categories alphabetically in a row
def sort_aspect_categories(row):
    aspects_list = row.split(", ")
    sorted_aspects = sorted(aspects_list)
    return ", ".join(sorted_aspects)

# Apply the function to the Aspect_Categories column
df['Sorted_Aspect_Categories'] = df['Aspect_Categories'].apply(sort_aspect_categories)

# Display the DataFrame with sorted aspect categories
print(df)
```

```python
# Save the modified DataFrame to a new CSV file
df.to_csv('sorted_data_modelling2.csv', index=False)
```
