```python
# Load spaCy model
nlp = spacy.load('en_core_web_sm')
```

```python
# Read preprocessed data
data = pd.read_csv('preprocessed_data.csv')
```

```python
# Sample text from the dataset
sample_text = data['Tokenized_Review'].iloc[0]  
```

```python
# Attributes dictionary
attributes = {
    'fabric': {
        'positive': ["soft", "breathable", "smooth", "silky", "luxurious", "plush", "comfortable", "cozy", "warm"],
        'neutral': ["lightweight", "stiff", "heavy", "airy", "thin", "thick", "absorbent","easily stretched"],
        'negative': ["itchy", "irritating", "scratchy", "durable", "terrible","seethrough","unsewn","overstretched"]
    },
    'design': {
        'positive': ["stylish", "elegant", "trendy", "unique", "modern", "beautiful", "fashionable", "cute", "pretty"],
        'neutral': ["simple", "classic", "plain", "basic", "standard", "ordinary", "subtle"],
        'negative': ["tacky", "ugly", "outdated", "unattractive", "boring", "poorly designed","flaw"]
    },
    'size': {
        'positive': ["perfect", "comfy", "flattering", "fitting", "roomy", "slim", "shapeless", "proportional", "oversized", "fit"],
        'neutral': ["loose", "tight", "bulky"],
        'negative': ["baggy", "inconsistent", "awkward", "oversized", "undersized", "snug", "long", "big", "tight", "small", "short"]
    },
    'color': {
        'positive': ["vibrant", "beautiful", "eye-catching", "rich", "vivid", "multicolored", "bright", "pastel","gorgeous","nice"],
        'neutral': ["neutral", "monochromatic", "subdued"],
        'negative': ["faded", "dull", "mismatched", "off-color", "true-to-color"]
    },
    'price': {
        'positive': ["affordable", "budget friendly", "reasonably", "bargain", "inexpensive", "economical"],
        'neutral': ["standard", "fair", "average", "typical"],
        'negative': ["expensive", "overpriced", "pricey", "costly", "steep", "premium"]
    },
    'comfort': {
        'positive': ["cozy", "soft", "breathable", "comfortable", "relaxing", "cushioned", "smooth", "comfy", "easy to wear", "supportive", "luxurious"],
        'neutral': ["adequate", "acceptable", "satisfactory", "standard", "tolerable"],
        'negative': ["restrictive", "itchy", "tight", "irritating", "scratchy"]
    },
    'quality': {
        'positive': ["excellent", "superior", "premium", "exceptional", "reliable", "top-notch", "outstanding", "durable", "high grade", "fine"],
        'neutral': ["average", "standard", "acceptable", "satisfactory", "typical"],
        'negative': ["poor", "inferior", "substandard", "low quality", "flimsy", "cheap", "faulty", "bad quality", "tacky", "deteriorating", "unsewn"]
    },
    'durability': {
        'positive': ["durable", "sturdy", "high-quality", "long-lasting", "lasting", "strong", "resilient", "robust", "solid"],
        'neutral': ["average", "typical", "standard", "acceptable"],
        'negative': ["low-quality", "flimsy", "easily-damaged", "wears-out-quickly", "fragile", "shoddy", "rugged", "tearing", "ripped"]
    },
    'service': {
        'positive': ["helpful", "courteous", "responsive", "professional"],
        'neutral': ["average", "standard", "satisfactory"],
        'negative': ["poor", "unprofessional", "slow", "disappoint"]
    },
}
```

```python
# Updated affirmation words
affirmation_words = [
    "good", "great", "excellent", "positive", "high",
    "amazing", "awesome", "fantastic", "superb", "wonderful",
    "pleasing", "satisfying", "delightful", "terrific", "impressive"
]

# Updated negation words
negation_words = [
    "not", "no", "never", "none", "neither", "low", "cheap", "too",
    "poor", "bad", "negative", "disappointing", "inferior",
    "unsatisfactory", "lacking", "subpar", "unimpressive", "terrible"
]
```

```python
# Create PhraseMatcher instances for clothing attributes
attribute_matchers = {key: PhraseMatcher(nlp.vocab) for key in attributes}
for key, labels in attributes.items():
    patterns = [nlp(text) for text in labels['positive'] + labels['neutral'] + labels['negative']]
    attribute_matchers[key].add(key, None, *patterns)
```

```python
# Function to find aspects and check for affirmation/negation
def find_attributes(text):
    doc = nlp(text)
    matched_aspects = set()  # Use a set to store unique aspects
    matched_categories = set()  # Use a set to store unique categories
    for key, matcher in attribute_matchers.items():
        matches = matcher(doc)
        for _, start, end in matches:
            aspect_span = doc[start:end]
            # Normalize aspect word to its base form (lemma)
            aspect = aspect_span.lemma_
            # Look for preceding and following tokens
            prev_token = doc[start - 1] if start > 0 else None
            next_token = doc[end] if end < len(doc) else None
            # Check for affirmation or negation words
            if prev_token and prev_token.lower_ in affirmation_words:
                aspect = prev_token.text + ' ' + aspect
            if next_token and next_token.lower_ in affirmation_words:
                aspect = aspect + ' ' + next_token.text
            # Store the aspect and its category
            matched_aspects.add(aspect)  # Use set.add() to add unique aspects
            matched_categories.add(key.capitalize())  # Store the category in title case
    return ', '.join(matched_aspects), ', '.join(matched_categories)
```

```python
# Convert NaN values to empty strings in the 'Tokenized_Review' column
data['Tokenized_Review'] = data['Tokenized_Review'].fillna('')

# Apply attribute matching to the dataset and add the 'Extracted_Aspects' and 'Aspect_Categories' columns
tqdm.pandas(desc="Matching Attributes")
data[['Extracted_Aspects', 'Aspect_Categories']] = data['Tokenized_Review'].astype(str).progress_apply(find_attributes).apply(pd.Series)
```

```python
# Generate word cloud for matched attributes
attributes_text = ' '.join(data['Extracted_Aspects'].explode().dropna())
wordcloud = WordCloud(width=800, height=400).generate(attributes_text)

# Display the word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
